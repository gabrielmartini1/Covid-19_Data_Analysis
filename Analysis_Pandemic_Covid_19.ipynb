##**1. Introdução**

<p align="justify">O Covid-19, é uma doença infecciosa causada pelo coronavírus denominado SARS-CoV-2, que foi identificado pela primeira vez em dezembro de 2019 em Wuhan, na região central da China.</p>

<p align="justify">Em 30 de janeiro de 2020, a Organização Mundial da Saúde (OMS) declarou que a epidemia da Covid-19 constituía uma Emergência de Saúde Pública de Importância Internacional, e, em 11 de março de 2020, uma pandemia.</p>

<p align="justify">O coronavírus pode ser transmitido por contato direto com uma pessoa infectada (durante um aperto de mão seguido do toque nos olhos, nariz ou boca), por meio da exposição a gotículas respiratórias expelidas, contendo o vírus, por uma pessoa infectada quando ela tosse ou espirra (com menos de 1 metro de distância da outra pessoa) e a transmissão por aerossol, que são as gotículas respiratórias menores, contendo o vírus e que podem permanecer suspensas no ar, serem carregadas por distâncias maiores que 1 metro e por períodos mais longos.</p> 

<p align="justify">Os sintomas do Covid-19 podem variar de leves a graves e incluem febre, tosse seca, fadiga, falta de ar, dor de cabeça, dor de garganta, perda de olfato ou paladar, entre outros. Algumas pessoas infectadas podem não apresentar sintomas, mas ainda assim serem capazes de transmitir o vírus para outras pessoas.</p>

<p align="justify">Embora as vacinas tenham sido desenvolvidas e a distribuição tenha começado em vários países, o Covid-19 ainda apresenta uma ameaça significativa à saude pública global e continua a desafiar governos e sistemas de saúde.</p>

<p align="justify">Os esforços para conter a propagação do vírus, desde o início da crise sanitária, incluem medidas como o distanciamento social, uso de máscaras, higiene adequada das mãos, quarentenas e, em último caso, lockdowns.</p>

<p align="justify">O primeiro caso de Covid-19 no Brasil, foi confirmado em 26 de fevereiro de 2020. Tratava-se de um homem idoso, residente de São Paulo, que havia retornado de uma viagem internacional da Itália. O primeiro óbito ocorreu em 17 de março de 2020.</p>

<p align="justify">No Brasil, o Covid-19 se espalhou rapidamente desde o início de 2020 e causou milhões de casos e mortalidades. O país enfrentou grandes desafios para conter a propagação do vírus, devido a uma série de fatores, incluindo desigualdades socioeconômicas, falta de recursos de saúde adequados em algumas regiões, atrasos na implementação de medidas preventivas e falta de coordenação em nível nacional.</p>

<p align="justify">Desde o início da pandemia, houve uma grande quantidade de dados sobre o Covid-19 sendo coletados e divulgados por governos, organizações internacionais, instituições de pesquisa, hospitais, laboratórios e outras fontes. Esses dados incluem o número de casos confirmados, mortes, recuperados, hospitalizados, quantidade de testes realizados, quantidade de vacinas administradas, entre outros.</p>

<p align="justify">Entretanto, houve uma série de questões e controvérsias relacionadas à coleta, análise e divulgação de dados sobre o Covid-19 no Brasil. Algumas das principais questões incluem:</p>

<p align="justify"> a) Falta de padronização: os dados sobre o Covid-19 são coletados por diversas agências governamentais e organizações em todo o país, e nem sempre existe uma padronização clara na forma como esses dados são coletados e relatados. Isso pode levar a inconsistências e discrepâncias nos dados.</p>

<p align="justify"> b) Subnotificação: há suspeitas de que os números oficiais de casos e mortes por Covid-19 no Brasil possam estar subestimados devido a uma série de fatores, incluindo falta de testagem adequada e subnotificação por parte das autoridades.</p>

<p align="justify"> c) Dificuldades na coleta de dados: o Brasil é um país enorme e diverso, com muitas áreas remotas e pobres que podem ter dificuldades em fornecer dados precisos e atualizados sobre a situação do Covid-19.</p>

<p align="justify"> d) Disputas políticas: a pandemia do Covid-19 se tornou uma questão altamente politizada no Brasil, e as autoridades em diferentes níveis do governo muitas vezes têm interesses conflitantes ou agendas políticas que podem afetar a coleta e divulgação de dados.</p>

<p align="justify">Os cientistas de dados têm um papel fundamental na análise e interpretação dos dados do Covid-19. Eles podem usar técnicas de análise de dados avançadas, como aprendizado de máquina e mineração de dados, para ajudar a entender a propagação da doença e entender seu impacto na saúde e segurança pública, e também na economia global.</p>

<p align="justify">Embora os dados do Covid-19 sejam uma ferramenta importante para combater a pandemia, é importante lembrar que eles nem sempre são precisos ou completos. Existem muitos fatores que podem afetar a coleta e a precisão dos dados, incluindo a disponibilidade de testes, a qualidade dos sistemas de relatórios de saúde e a falta de transparência em alguns países.</p>

<p align="justify">No entanto, os dados do Covid-19 continuam a ser uma ferramenta essencial para entender e combater a pandemia em todo o mundo. ###**3.1 Nomenclatura das variáveis**
<br>

**Casos Confirmados (*Confirmed cases*)**
* `total_cases` - Total de casos confirmados de COVID-19.
* `new_cases` - Novos casos confirmados.
* `new_cases_smoothed` - Novos casos confirmados (com média móvel de 7 dias).
* `total_cases_per_million` - Total de casos confirmados por 1 milhão de pessoas.
* `new_cases_per_million` - Novos casos por 1 milhão de pessoas.
* `new_cases_smoothed_per_million` - Novos casos confirmados (com média móvel de 7 dias) por 1 milhão de pessoas.

<br>

**Mortes confirmadas (*Confirmed deaths*)**
* `total_deaths` - Total de mortes atribuídas à COVID-19.
* `new_deaths` - Novas mortes atribuídas à COVID-19.
* `new_deaths_smoothed` - Novas mortes atribuídas à COVID-19 (com média móvel de 7 dias).
* `total_deaths_per_million` - Total de mortes por 1 milhão de pessoas.
* `new_deaths_per_million` - Novas mortes atribuídas à COVID-19 por 1 milhão de pessoas.
* `new_deaths_smoothed_per_million` - Novas mortes atribuídas à COVID-19 (com média móvel de 7 dias) por 1 milhão de pessoas.

<br>

**Excesso de mortalidade (*Excess mortality*)**
* `excess_mortality` - Diferença percentual entre o número relatado de mortes semanais ou mensais em 2020/2021 e o número projetado de mortes para o mesmo período com base nos anos anteriores. 
* `excess_mortality_cumulative` - Diferença percentual entre o número acumulado de mortes desde 1º de janeiro de 2020 e as mortes projetadas cumulativas para o mesmo período com base nos anos anteriores.
* `excess_mortality_cumulative_absolute` - Diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores.
* `excess_mortality_cumulative_per_million` - Diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores, por milhão de pessoas.

<br>

**Hospital e UTI (*Hospital & ICU*)**
* `icu_patients` - Número de pacientes com COVID-19 em unidades de terapia intensiva (UTIs) em um determinado dia. 
* `icu_patients_per_million` - Número de pacientes com COVID-19 em unidades de terapia intensiva (UTIs) em um determinado dia por 1 milhão de pessoas.
* `hosp_patients` - Número de pacientes com COVID-19 internados em um determinado dia.
* `hosp_patients_per_million` - Número de pacientes com COVID-19 hospitalizados em um determinado dia por 1 milhão de pessoas.
* `weekly_icu_admissions` - Número de pacientes com COVID-19 recém-admitidos em unidades de terapia intensiva (UTIs) em uma determinada semana.
* `weekly_icu_admissions_per_million` - Número de pacientes com COVID-19 recém-admitidos em unidades de terapia intensiva (UTIs) em uma determinada semana por 1 milhão de pessoas.
* `weekly_hosp_admissions` - Número de pacientes com COVID-19 recém-admitidos em hospitais em uma determinada semana.
* `weekly_hosp_admissions_per_million` - Número de pacientes com COVID-19 recém-admitidos em hospitais em uma determinada semana por 1 milhão de pessoas.

<br>

**Respostas de políticas (*Policy responses*)**
* `stringency_index` - Índice de Severidade de Resposta do Governo: medida composta com base em 9 indicadores de resposta, incluindo fechamento de escolas, fechamento de locais de trabalho e proibições de viagens, reescalonados para um valor de 0 a 100, onde o valor de 100 é a resposta mais rigorosa.

<br>

**Taxa de reprodução (*Reproduction rate*)**
* `reproduction_rate` - Estimativa em tempo real da taxa efetiva de reprodução (R) do COVID-19.

<br>

**Testes e positividade (*Tests & positivity*)**
* `total_tests` - Total de testes para COVID-19
* `new_tests` - Novos testes para COVID-19 (calculados apenas para dias consecutivos)
* `total_tests_per_thousand` - Total de testes para COVID-19 por 1.000 pessoas.
* `new_tests_per_thousand` - Novos testes para COVID-19 por 1.000 pessoas.
* `new_tests_smoothed` - Novos testes para COVID-19 (com média móvel de 7 dias).
* `new_tests_smoothed_per_thousand` - Novos testes para COVID-19 (com média móvel de 7 dias) por 1.000 pessoas.
* `positive_rate` - A parcela de testes positivos para COVID-19, dada como uma média contínua de 7 dias.
* `tests_per_case` - Testes realizados por novo caso confirmado de COVID-19, dados como uma média contínua de 7 dias.
* `tests_units` - Unidades usadas pelo local para relatar seus dados de teste.

<br>

**Vacinações (*Vaccinations*)**
* `total_vaccinations` - Número total de doses de vacinação contra COVID-19 administradas.
* `people_vaccinated` - Número total de pessoas que receberam pelo menos uma dose de vacina.
* `people_fully_vaccinated` - Número total de pessoas que receberam todas as doses de vacinação.
* `total_boosters` - Número total de doses de reforço da vacinação administradas contra a COVID-19.
* `new_vaccinations` - Novas doses de vacinação administradas contra a COVID-19.
* `new_vaccinations_smoothed` - Novas doses de vacinação COVID-19 administradas (com média móvel de 7 dias). 
* `total_vaccinations_per_hundred` - Número total de doses de vacinação contra COVID-19 administradas por 100 pessoas na população total.
* `people_vaccinated_per_hundred` - Número total de pessoas que receberam pelo menos uma dose de vacina por 100 pessoas na população total.
* `people_fully_vaccinated_per_hundred` - Número total de pessoas que receberam todas as doses prescritas pelo protocolo inicial de vacinação por 100 pessoas na população total.
* `total_boosters_per_hundred` - Número total de doses de reforço da vacinação contra COVID-19 administradas por 100 pessoas na população total.
* `new_vaccinations_smoothed_per_million` - Novas doses de vacinação contra COVID-19 administradas (com média móvel de 7 dias) por 1 milhão de pessoas na população total.
* `new_people_vaccinated_smoothed` - Número diário de pessoas que receberam a primeira dose da vacina (com média móvel de 7 dias).
* `new_people_vaccinated_smoothed_per_hundred` - Número diário de pessoas que receberam a primeira dose da vacina (com média móvel de 7 dias) por 100 pessoas na população total.

###**3.5 Teoria do IQR Score**

Para definirmos o IQR Score (Intervalor Interquartil) de qualquer variável, será necessário calcular anteriormente o seu quartil 1 (Q1) e o quartil 3 (Q3).

Supondo que estamos calculando o IQR da variável 'total_cases', logo temos:

* Q1 = df['total_cases'].quantile(0.25)
* Q3 = df['total_cases'].quantile(0.75)
* Portanto, IQR = Q3 - Q1

Para calcularmos o limite inferior e superior, temos:<br>
* limite_inferior  = Q1 - 1.5 * IQR<br>
* limite_superior = Q3 + 1.5 * IQR

Com o resultado numérico do IQR Score, será possível encontrarmos os limites superior e inferior para cada variável. Sabendo que valores acima dos limites encontrados serão considerados outliers, logo poderão ser excluídos do dataset. 

Para trazer mais clareza ao assunto do IQR Score, vamos visualizar um exemplo do diagrama de caixa com intervalo interquartil, conforme a figura 1 abaixo:

```python
# Cálculo do Limite Superior -  p/ a variável 'total_cases'
limite_superior = df_clean['total_cases'].quantile(0.75) + 1.5 * (df_clean['total_cases'].quantile(0.75) - df_clean['total_cases'].quantile(0.25))

# Cálculo do Limite Inferior -  p/ a variável 'total_cases'
limite_inferior = df_clean['total_cases'].quantile(0.25) - 1.5 * (df_clean['total_cases'].quantile(0.75) - df_clean['total_cases'].quantile(0.25))

print(f'Limite superior de total_cases: {limite_superior:.2f}')
print(f'Limite inferior de total_cases: {limite_inferior:.2f}\n')

# Plotar o boxplot da variável 'total_cases'
df_clean.total_cases.plot(kind="box", vert=False, figsize=(15, 3), color = 'blue')
plt.show()
print('\n')

# Visualizar o total de entradas em (%) acima do limite superior  
print(f'Acima do limite superior de 1972497.00 ===>  ', (len(df_clean[df_clean.total_cases > 1972497.00])), 'entradas',
      "({:.2f}%)".format((len(df_clean[df_clean.total_cases > 1972497.00]) / df_clean.shape[0])*100))

# Visualizar o total de entradas em (%) até o limite superior
print(f'Até o limite superior de 1972497.00 ======> ', (len(df_clean[df_clean.total_cases <= 1972497.00])),'entradas',
      "({:.2f}%)".format((len(df_clean[df_clean.total_cases <= 1972497.00]) / df_clean.shape[0])*100))

# Calcular o coeficiente de assimetria (skewness) da variável 'total_cases'
skewness = skew(df_clean['total_cases'])
print("Coeficiente de assimetria (skewness) =====>   {:.2f}".format(skewness))
```

```python
fig, ax = plt.subplots(figsize=(7, 4))

ax.bar(media_stringency.index, media_stringency.values, color='brown', width=0.5)
ax.set_xticklabels(media_stringency.index, rotation=360, ha='center')
ax.set_ylim(0, 100)
ax.set_xlabel('')
ax.set_ylabel('Índice de Severidade')
ax.set_title('Média do Índice de Severidade por País até 20/03/2023')

plt.show()

```
